# Prediction of Good Creditors' Limit Bal Only

```python
good_creditors = rawData[rawData['loan paid'] == 1]
```

```python
X_good = good_creditors.iloc[:, 2:-1]
print('Summary of feature sample')
X_good.head()
```

```python
#dependent variable
y_good = good_creditors['LIMIT_BAL']
y_good.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_good, y_good, test_size=0.25, random_state=123)
```

```python
#Modeling (Classification)
algo = DecisionTreeClassifier()
model = algo.fit(X_train,y_train)

#Predictions
preds = model.predict(X_test)

print(classification_report(y_test, preds))
```

# Prediction of all Customers' default status

```python
X = rawData.iloc[:, 1:25]
print('Summary of feature sample')
X.head()
```
